# Import the packages

```python
import pandas as pd 


import plotly as py

from plotly.offline import init_notebook_mode, iplot
import plotly.graph_objs as go

import os
init_notebook_mode(connected=True)
```

```python
# Load data that we will use.
timesData = pd.read_csv("timesData.csv")
```

# Line Plot
go.Scatter method takes in parameters like

x : The data to be plotted in x axis

y : The data to be plotted in y axis

mode : since its a line chart we set the mode as lines

marker : we can use this to set the color.It should be a dictionary

data : a list with the taces as its elments.

layout : This can be used to set the title of the graph,x axis,y axis.It should be a dictionary.

fig : a dictionary with keys data and layout.

iplot : used to plot the graph takes in a dictionary fig.

```python
# prepare data frame
df = timesData.iloc[:100,:]

# Creating trace1
trace1 = go.Scatter(
                    x = df.world_rank,
                    y = df.citations,
                    mode = "lines",
                    name = "Citations",
                    marker = dict(color = 'rgb(16, 112, 2)'),
                    text= df.university_name)

# Creating trace2
trace2 = go.Scatter(
                    x = df.world_rank,
                    y = df.teaching,
                    mode = "lines",
                    name = "Teaching",
                    marker = dict(color = 'rgb(80, 26, 80)'),
                    text= df.university_name)
data = [trace1, trace2]
layout = dict(title = 'Citation and Teaching vs World Rank of Top 100 Universities',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
             )

fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Scatter plot

We use go.Scatter method

Here we change the mode to markers 

```python
# prepare data frames
df2015 = timesData[timesData.year == 2015].iloc[:100,:]
df2016 = timesData[timesData.year == 2016].iloc[:100,:]


# creating trace1
trace1 =go.Scatter(
                    x = df2015.world_rank,
                    y = df2015.citations,
                    mode = "markers",
                    name = "2015",
                    marker = dict(color = 'rgb(255, 128, 2)'),
                    text= df2015.university_name)
# creating trace2
trace2 =go.Scatter(
                    x = df2016.world_rank,
                    y = df2016.citations,
                    mode = "markers",
                    name = "2016",
                    marker = dict(color = 'rgb(0, 255, 200)'),
                    text= df2016.university_name)

data = [trace1, trace2]
layout = dict(title = 'Citation vs world rank of top 100 universities with 2014, 2015 and 2016 years',
              xaxis= dict(title= 'World Rank',ticklen= 5,zeroline= False),
              yaxis= dict(title= 'Citation',ticklen= 5,zeroline= False)
             )
fig = go.Figure(data = data, layout = layout)
iplot(fig)
```

# Bar Plot
We use go.Bar method

```python
# prepare data frames
df2014 = timesData[timesData.year == 2014].iloc[:3,:]
# import graph objects as "go"
import plotly.graph_objs as go
# create trace1 
trace1 = go.Bar(
                x = df2014.university_name,
                y = df2014.citations,
                name = "citations",
                marker = dict(color = 'rgb(251,140,0)'),
                text = df2014.country)
# create trace2 
trace2 = go.Bar(
                x = df2014.university_name,
                y = df2014.teaching,
                name = "teaching",
                marker = dict(color = 'rgb(25,118,210)'),
                text = df2014.country)
data = [trace1, trace2]
layout = dict(barmode = "group")
fig = go.Figure(data = data, layout = layout)
iplot(fig)

```

# Histogram

Here we need to use go.Histogram method.

```python
# prepare data
x2011 = timesData.student_staff_ratio[timesData.year == 2011]
x2012 = timesData.student_staff_ratio[timesData.year == 2012]

trace1 = go.Histogram(
    x=x2011,
    opacity=0.75,
    name = "2011",
    marker=dict(color='rgb(171, 50, 96)'))

trace2 = go.Histogram(
    x=x2012,
    opacity=0.75,
    name = "2012",
    marker=dict(color='rgb(12, 50, 196)'))

data = [trace1, trace2]
layout = dict(barmode='overlay',
                   title=' students-staff ratio in 2011 and 2012',
                   xaxis=dict(title='students-staff ratio'),
                   yaxis=dict( title='Count'),
)
fig = go.Figure(data=data, layout=layout)
iplot(fig)
```

offline.plot method can be used to save the plots offline as a HTML file.
Takes in the fig to be saved and the name of the html file.

```python
py.offline.plot(fig, filename='student-staff-ratio.html');
```

# Box Plots

We use go.Box for Box plots.

```python
# data preparation
x2015 = timesData[timesData.year == 2015]

trace1 = go.Box(
    y=x2015.total_score,
    name = 'total score of universities in 2015',
    marker = dict(
        color = 'rgb(12, 12, 140)',
    )
)
trace2 = go.Box(
    y=x2015.research,
    name = 'research of universities in 2015',
    marker = dict(
        color = 'rgb(12, 128, 128)',
    )
)
data = [trace1, trace2]
fig = go.Figure(data=data)
iplot(data)
```
